```python
from urllib.request import urlopen
```

```python
f = urlopen("https://www.goodreads.com/book/show/4671.The_Great_Gatsby")
```

```python
html = str(f.read())
```

```python
html
```

```python
reviws = html.split('<div id="review_')[1:]
```

```python
len(reviws)
```

```python
reviws[0]
```

```python
def parse_review(review):
    d = {}
    d['stars'] = review.split('<span class=" staticStars notranslate" title="')[1].split('"')[0]
    d['date'] = review.split('<a class="reviewDate')[1].split('>')[1].split('<')[0]
    d['user'] = review.split('<a title')[1].split('""')[0]
    
    shelves = []
    try:
        shelf_Blook = review.split('<div class="uitext greyText bookshelves">')[1].split('</div')[0]
        for s in shelf_Blook.split('shelf=')[1:]:
            shelves.append(s.split('""')[0])
        d['shelves'] = shelves
    except Exception as e:
        pass
    review_block = review.split('<div class="reviewText stacked">')[1].split('</div')[0]
    d['reviewBlock'] = review_block
    return d
```

```python
review_dict = [parse_review(r) for r in reviws]
review_dict[0]
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(review_dict[0]['reviewBlock'])
```
